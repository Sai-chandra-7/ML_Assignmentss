Task 2

```python
# drop null target rows
df = df.dropna(subset = ['selling_price'])
```

```python
# impute missing input features
df = df.fillna(df.median(numeric_only = True))
```

```python
# strip and lowercase the brand column
df['brand'] = df['brand'].str.strip().str.lower()
```

```python
#extract numeric values from the mileage column
df['mileage'] = df['mileage'].str.replace('kmpl','').str.replace('km/kg','')
df['mileage'] = df['mileage'].astype(float)
```

```python
#remove duplicates
df = df.drop_duplicates()
```

Task 3

```python
mean_price = df['selling_price'].mean()
```

```python
df['predicted_price'] = mean_price
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(df['selling_price'] , df['predicted_price'])
print("Baseline MAE :" , mae)
```
