```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['city'] = df['city'].fillna(df['city'].mode()[0])
```

```python
df = df.drop_duplicates()
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# one hot encoding to city
df = pd.get_dummies(df, columns=['city'])

#label encoding to gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

features = ['age', 'annual_income']

# Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[features] = minmax.fit_transform(df[features])
print("Min-Max Scaled:\n", df_minmax[features])

# Standardisation
standard = StandardScaler()
df_standard = df.copy()
df_standard[features] = standard.fit_transform(df[features])
print("\nStandardised:\n", df_standard[features])
```
